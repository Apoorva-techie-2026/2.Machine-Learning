```python
# Importing libraries from the database
import pandas as pd
```

```python
#Loading the dataset
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
#Splitting the input and output with the variable name
independent=dataset[['YearsExperience']]
independent
dependent=dataset[['Salary']]
dependent
```

```python
#Splitting the train set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
weight=regressor.coef_
baise=regressor.intercept_
```

```python
#Evaluating the values
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#Saving the model
import pickle
filename="SLR_Depolyment.sav"
pickle.dump(regressor,open(filename,'wb')) 
```

```python
#Checkimg the model is working correctly in real time(Deployment)
loaded_model=pickle.load(open("SLR_Depolyment.sav",'rb'))
result=loaded_model.predict([[15]])
result
```
